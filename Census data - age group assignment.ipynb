```python
def set_age_group(row):
    if row['maximum_age']<= 14.0:
        return "Children"
    elif row['maximum_age']<= 24.0:
        return "Youth"
    elif row['maximum_age']<= 64.0:
        return "Adults"
    elif pd.isnull(row.maximum_age):
        return "Unknown"
    else:
        return "Seniors"

census2010 = census2010.assign(age_group=census2010.apply(set_age_group, axis=1))

```
